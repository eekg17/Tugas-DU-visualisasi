```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

Makanan_penutup = ('Donat','Pastel','Kue Coklat','Ice Cream','Puding Vanila',
                   'Brownies','Puding Strawberi','Puding Coklat','Ice Cream Nutela',
                    'Kue Coklat-Keju','Kue Wajik','Kue Sus','Mochi')
banyak = (14, 5, 12, 19, 6, 8, 12, 9, 10, 17, 2, 9, 13)

df = pd.DataFrame({'Penjualan': Makanan_penutup, 'Qty': banyak})
df.sort_values(by='Qty', inplace = True, ascending = False)

kor_x = np.arange(len(df))
warna = ['#000080' for _ in range(len(df))]
warna [-3] = '#FF0000'
warna [-2] = '#FF0000'
warna [-1] = '#FF0000'

plt.title('Menu Kafe Biru\n')
plt.bar(kor_x, df['Qty'],tick_label=df['Penjualan'],color = warna)
plt.xticks(rotation=90)
plt.ylabel('Range Makanan Kafe Biru\n')
plt.xlabel('\nBerikut Menu Makanan Penutup')
plt.show()
```

saya memilih bar chart karna dapat menyajikan grafik garis serta dapat menampilkan data pertumbuhan dan dapat melihat perbandinganya
